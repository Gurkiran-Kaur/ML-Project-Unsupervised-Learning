```python
# Part I : EDA - Exploratory Data Analysis & Pre-processing
# 1.1 Data Import.

import pandas as pd

# Load the dataset into dataframe
df_grocery = pd.read_csv('Wholesale_Data.csv')

import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')
```

```python
# 1.2 Data description and cleaning
# 1.2.1 Data Description: Generate summary statistics for each column of the dataset.

# Generate summary statistics
summary_statistics = df_grocery.describe()

# Print the summary statistics
print(summary_statistics)
```

```python
# 1.2.2 Data Cleaning: Check the dataset for any missing or incorrect data and clean the dataset accordingly.

# Step 1: Check for missing data
missing_data = df_grocery.isnull().sum()
print("Missing Data:")
print(missing_data)

# Step 2: Handle missing data (if any), Impute missing values with the column mean
df_grocery.fillna(df_grocery.mean(), inplace=True)

# Step 3: Check for incorrect data (if any), Check for negative values in columns
columns_incorrect = ['Channel', 'Region', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
negative = df_grocery[df_grocery[columns_incorrect] < 0]
print("\nIncorrect Data (Negative):")
print(negative)

# Step 4: Handle incorrect data (if any), Remove rows with negative values in columns
df_grocery = df_grocery[df_grocery[columns_incorrect] >= 0]

# After cleaning, re-check for missing data to ensure it's all cleaned up
missing_data_after_cleaning = df_grocery.isnull().sum()
print("\nMissing Data after Cleaning:")
print(missing_data_after_cleaning)

df_grocery.head(25)
```

```python
# 1.3 Data Description: To understand and cross verify the transformation performed in previous steps.

# Generate summary statistics
summary_statistics = df_grocery.describe()

# Print the summary statistics
print(summary_statistics)
```

```python
# 1.4 Data Visualization: Create visualizations such as histograms, box plots, scatter plots, and heatmaps to understand the relationships and trends between the different variables in the dataset

import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical variables
df_grocery[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].hist(figsize=(12, 8))
plt.suptitle('Histograms of Variables')
plt.show()

# Box plots for numerical variables
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_grocery[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']])
plt.title('Box Plots of Variables')
plt.show()

# Scatter plot (example: Fresh vs. Milk)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_grocery, x='Fresh', y='Milk', hue='Channel')
plt.title('Scatter Plot: Fresh vs. Milk')
plt.show()

# Heatmap (correlation matrix)
plt.figure(figsize=(10, 8))
corr_matrix = df_grocery.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 1.4.1 Outlier Detection: Check for any outliers in the dataset and determine whether they are valid or erroneous data points.

# Define the numerical columns for which you want to detect outliers
numerical_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Function to detect outliers using IQR method
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

# Detect outliers for each numerical column
outliers = df_grocery[numerical_columns].apply(detect_outliers_iqr)

# Print the number of outliers for each column
print("Number of Outliers:")
print(outliers.sum())

# Print the rows containing outliers
print("\nRows with Outliers:")
print(df_grocery[outliers.any(axis=1)])
```

```python
# 1.4.2 Correlation Analysis: Calculate the correlation between different variables in the dataset to determine which variables are highly correlated and which ones are not.

# Calculate the correlation matrix
correlation_matrix = df_grocery.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# You can also visualize the correlation matrix using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Explanation:

This code first calculates the correlation matrix using df_grocery.corr().
It then prints the correlation matrix, showing how each variable is correlated with every other variable.
Additionally, it visualizes the correlation matrix as a heatmap for better understanding.
Interpreting the Correlation Matrix:

The values in the matrix range from -1 to 1.

1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
0 indicates no correlation.
Variables with high positive correlation (close to 1) are positively associated. Changes in one variable are likely to be associated with changes in the other variable in the same direction.

Variables with high negative correlation (close to -1) are negatively associated. Changes in one variable are likely to be associated with changes in the other variable in the opposite direction.

Variables with low correlation (close to 0) have little linear relationship with each other.

Keep in mind that correlation does not imply causation. It only measures the strength of a linear relationship between two variables.

```python
# 1.5 Data Transformation: Transform the data by standardizing or normalizing the variables to make them comparable across different scales.

from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define the numerical columns for transformation
numerical_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Standardization (Z-score scaling)
scaler_standardization = StandardScaler()
df_grocery_standardized = pd.DataFrame(scaler_standardization.fit_transform(df_grocery[numerical_columns]), columns=numerical_columns)

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_grocery_minmax = pd.DataFrame(scaler_minmax.fit_transform(df_grocery[numerical_columns]), columns=numerical_columns)

# Print the first few rows of the standardized DataFrame
print("Standardized Data:")
print(df_grocery_standardized.head())

# Print the first few rows of the min-max scaled DataFrame
print("\nMin-Max Scaled Data:")
print(df_grocery_minmax.head())
```

```python
# 1.6 Feature Selection: Identify the most important features or variables that contribute the most to the overall variance in the dataset.

from sklearn.ensemble import RandomForestRegressor

# Define the features (exclude non-numeric columns like Channel and Region)
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Create a DataFrame with only the features
X = df_grocery[features]

# Standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
from sklearn.decomposition import PCA
pca = PCA()
principal_components = pca.fit_transform(X_scaled)

# Calculate the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Print the explained variance ratio for each principal component
print("Explained Variance Ratio (PCA):")
print(explained_variance_ratio)

# Visualize explained variance ratio for PCA
import matplotlib.pyplot as plt
plt.bar(range(1, len(explained_variance_ratio)+1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio for Each Principal Component (PCA)')
plt.show()

# Apply Random Forest Regression for feature importance
y = df_grocery['Channel']  # 'Channel' is the target variable

# Create a random forest regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_scaled, y)

# Get feature importance scores
feature_importance = rf_regressor.feature_importances_

# Print feature importance scores
print("\nFeature Importance (Random Forest Regression):")
for feature, importance in zip(features, feature_importance):
    print(f"{feature}: {importance}")
```

```python
# Part II - KMeans Clustering

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set environment variable to avoid memory leak warning on Windows with MKL
os.environ['OMP_NUM_THREADS'] = '2'


# Define the features for clustering
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
X = df_grocery[features]

# Step 1: Determine the Optimal Number of Clusters (Elbow Method)
inertia_values = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)  # Set n_init explicitly
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Define the features for clustering
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
X = df_grocery[features]

# Plot the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()

# Based on the Elbow Method, choose an optimal number of clusters
k = 3  # Choosing 3 clusters based on the Elbow Method

# Step 2: Initialize the KMeans Algorithm with the chosen number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)

# Step 3: Standardize the Features (Important for KMeans)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Fit the Model to the Standardized Data
kmeans.fit(X_scaled)

# Step 5: Assign Data Points to Clusters
df_grocery['Cluster'] = kmeans.labels_

# Step 6 and 7: Update the Centroids and Repeat Until Convergence
# KMeans algorithm inherently handles this during fitting.

# Print the count of data points in each cluster
print("\nCount of Data Points in Each Cluster:")
print(df_grocery['Cluster'].value_counts())

# Print the centroids of the clusters (in original feature space)
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
centroids_df = pd.DataFrame(centroids, columns=features)
print("\nCluster Centroids:")
print(centroids_df)

# Visualize the clustering using a pairplot
sns.set(style="ticks")
sns.pairplot(df_grocery, hue='Cluster', palette='viridis')
plt.show()
```

```python
# Part III - Hierarchical Clustering

from scipy.cluster.hierarchy import dendrogram, linkage

# Define the features for clustering
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
X = df_grocery[features]

# Perform hierarchical clustering
linked = linkage(X, method='ward', metric='euclidean')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```
